## Question 2, 3: Testing Accuracy and Prediction Issues
When doing any sort of classification task, it is crucial to evaluate your performance as clearly as you can. To do this, we can run the helper functions that give us this output after fitting the classifer.

This section will also explain how the flow of the program works.

`bayes.get_newsgroups()` is a function that loads the newsgroups dataset from scikit's built-in system. The returned objects hold the raw data, and the data's class labels. The first document is printed below:


```python
twenty_train, twenty_test = bayes.get_newsgroups()
print('-----------')
print('Document 1 class: ' + str(twenty_train.target_names[twenty_train.target[0]]))
print('-----------')
print(twenty_train.data[0])
```

I then vectorize the data into both a bag-of-words model and a TFIDF model using scikits CountVectorizer() and TFiDFVectorizer().

```python
train_bow, test_bow, cv_bow = bayes.vectorize(twenty_train.data, twenty_test.data, model='bow')
train_tfidf, test_tfidf, cv_tfidf = bayes.vectorize(twenty_train.data, twenty_test.data, model='tfidf')
```

We then estimate the prior classes via MLE, the features give classes via MAP (with dirchlet prior, though I will compare it with Laplacian smoothing), and then predict values. 

Each of these operations is highly vectorized using the Numpy package, so where possible explict python for loops are avoided and operations are cast as vector or matrix mutiplications or elementwise operations. Each numpy array is structred to make it easy to perform these operations and they are blazingly fast.

```python
class_priors = bayes.phat_class_est(twenty_train.target, 
                                    twenty_train.target_names, 
                                    debug=False)

phat_words = bayes.phat_word_est(train_bow,
                                 labels=twenty_train.target
                                 #alpha=1
                                 )

predicted = bayes.predict(test_data=test_bow,
              test_labels=twenty_test.target,
              p_classes=class_priors,
              p_features=phat_words
              )
```

Now for the report:

```python
np.set_printoptions(precision=4)
rep = utils.report(predicted, twenty_test.target_names, print_report=True, print_cm=True)
```

The classifier does well, hitting 77% accuracy. It has a hard time with two clusters of documents, the comp.*
and the religious/political groups. These groups have a great deal of overlap in subjects so this is not too surprising.

## Question 4: Priors and Overfitting

The following blocks of code create our testing setup and fit the model with many  $\beta$ values.

```python
overfit_trials = [1/10**x for x in reversed(range(10))]
print(overfit_trials)
```

```python
%%capture
# capture magic command prevents printing by capturing stout
overfit_results = []
for beta in overfit_trials:
    phat_c,phat_w, pred, rep = bayes.run_model(twenty_train, twenty_test,
                beta=beta,
                bow=(train_bow, test_bow),
                report=False)
    overfit_results.append(rep[1])
```

```python
fit_results = np.array((overfit_trials, overfit_results))
pyplot.figure()
plt.plot(fit_results[0], fit_results[1])
# ax = fig.add_subplot(111)
plt.xscale('log')
plt.title('Effect of Beta choice on classifier accuracy')
plt.ylabel('classification accuracy')
plt.xlabel('beta')
plt.show()
```

As you can see, the choice of beta slightly changes the classification score, with $\beta = 0.1$ giving us our highest accuracy.

$$P(X_i|Y_k) = \frac{(count\ of\ X_i\ in\ Y_k)+(\alpha -1)}{(total\ words\ in Y_k)+((\alpha - 1)*(length\ of\ vocab\ list)))}$$

Where $$\alpha = 1 + \beta$$

tells us that $\beta$ weights the amount of influence the size of the vocabulary has over the estimated posterior probability. 

## Question 5,6: Important Features
If we multiply the estimated word priors and the class priors, we can see words that were important in our dataset. This is a sort-of indirect estimate of 
$P(C|W) = {P(W|C) * P(C)}$ This can allow us to pull the top words per class or the top words overall by looking at words per class. 

```python
# numpy will broadcast the smaller array along the larger array
# and we transpose it to get it back
# see http://wiki.scipy.org/EricsBroadcastingDoc
log_posteriors =  np.transpose(np.log2(phat_c) + np.log2(phat_w.T))
best_words = utils.top_words(cv_bow, log_posteriors,
                twenty_test.target_names,
                n=100, per_class=False, order='high')
worst_words = utils.top_words(cv_bow, log_posteriors,
                twenty_test.target_names,
                n=100, per_class=False, order='low')
best = [utils.get_word_count(cv_bow, train_bow, word) 
                        for word in best_words]
worst = [utils.get_word_count(cv_bow, train_bow, word) 
                         for word in worst_words]
```

```python
print(best)
```

Notice how the words that come out all have moderate counts and tent to be loaded - car, file, mac, government, jesus - likely high information words. Contrast those with the lowest info words in the set:

```python
print(worst)
```

That have very, very low counts in the data, are often mispelled.
